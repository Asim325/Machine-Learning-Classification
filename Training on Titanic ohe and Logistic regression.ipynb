```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("titanic.csv")
data
```

```python
data.info()
```

# One Hot Encoding

```python
real_data=data[["Sex","Embarked"]]
real_data
```

```python
dummies_df=pd.get_dummies(real_data)
dummies_df
```

```python
#assign dummies to variable
x=dummies_df
x
```

```python
y=data[['Survived']]
y
```

# Splitting data into train and test 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

# Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

# Logistic Regression Trainig

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
#on training set
classifier.score(x_train,y_train)
```

```python
#on test set
classifier.score(x_test,y_test)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
probab=classifier.predict_proba(x_test)
probab
```

```python
np.argmax(probab[3])
```

```python
plt.scatter(x_train,y_train , marker = 'x' , color = 'black' )
plt.plot(x_train,classifier.predict(x_train),color='cyan')

font1 = {'family':'serif','color':'black','size':20}
font2 = {'family':'serif','color':'green','size':18}

plt.xlabel('sex','embaraked',fontdict = font2)
plt.ylabel('Survived',fontdict = font2)
plt.title('Titanic Survival Classification',fontdict = font1,)
plt.show()
```
